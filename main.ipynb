```python
def calculate_age(dob):
    # Convert the date of birth string to a datetime object
    dob=str(dob)
    dob_date = datetime.strptime(dob, '%d-%b-%y')
    
    # Get the current date
    current_date = datetime.now()
    
    # Calculate the age
    
    x=dob_date.year
    y=current_date.year
    if(x>y):
        x=x-100
    
    age = y - x - ((current_date.month, current_date.day) < (dob_date.month, dob_date.day))
    
    return age
```

```python
def convert_height_to_cm(height):
    # Split the height string into feet and inches
    height=str(height)
    feet, inches =  height.split("'")
    inches=inches[:-1]
    # Convert feet and inches to total inches
    total_inches = int(feet) * 12 + int(inches)

    # Convert total inches to centimeters (1 inch = 2.54 cm)
    cm = total_inches * 2.54

    return cm
```
